# Dry Bean Dataset.
## Xây dựng mô hình phân loại các loại hạt trong bộ dữ liệu Dry Bean Dataset bằng hai thuật toán:

Decision Tree.
Random Forest.

## Dữ liệu
Dữ liệu đã được preprocessing (file processing_seeds.ipynb) và chia thành hai tập:

Dry_Bean_Dataset/
├── dry_bean_train.csv
└── dry_bean_test.csv

## Các bạn có thể tự tạo file processing để có thể processing theo ý mình

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import(
    accuracy_score,
    classification_report,
    confusion_matrix
)
```

```python
train = pd.read.csv("Dry_Bean_Dataset/dry_bean_train.csv")
test = pd.read.csv("Dry_Bean_Dataset/dry_bean_test.csv")
train.head()
train.shape
train.columns

# Chia X và y
X_train = train.drop("class", axis=1)
y_train = train["class"]
X_test = test.drop("class", axis=1)
y_test = test["class"]
print(X_train.shape)
print(X_test.shape)
```

## Huấn luyện Decision Tree.

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)
dt_model.fit(X_train, y_train)

# Dự đoán.
dt_pred = dt_model.predict(X_test)

# Accuracy.
dt_acc = accuracy_score(y_test, dt_pred)
print("Accuracy của Decision Tree là:", dt_acc)
print(classification_report(y_test, dt_pred))
```

## Confusion Matrix của Decision Tree.

```python
cm = confusion_matrix(y_test, dt_pred)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)
plt.title("Confusion Matrix của Decision Tree")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## Random Forest.

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf_model.fit(X_train, y_train)

# Dự đoán.
rf_pred = dt_model.predict(X_test)

# Accuracy.
rf_acc = accuracy_score(y_test, dt_pred)
print("Accuracy của Random Forest là:", rf_acc)
print(classification_report(y_test, rf_pred))
```

## Confusion Matrix của Random Forest.

```python
cm = confusion_matrix(y_test, rf_pred)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)
plt.title("Confusion Matrix của Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
