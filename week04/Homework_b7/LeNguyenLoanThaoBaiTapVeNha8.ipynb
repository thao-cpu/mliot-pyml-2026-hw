# Dry Bean Dataset.
## Xây dựng mô hình phân loại các loại hạt trong bộ dữ liệu Dry Bean Dataset bằng hai thuật toán:

Decision Tree.
Random Forest.

## Dữ liệu
Dữ liệu đã được preprocessing (file processing_seeds.ipynb) và chia thành hai tập:

Dry_Bean_Dataset/
├── dry_bean_train.csv
└── dry_bean_test.csv

## Các bạn có thể tự tạo file processing để có thể processing theo ý mình

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import(
    accuracy_score,
    classification_report,
    confusion_matrix
)
```

```python
train_df = pd.read.csv("Homework_b7/Dry_Bean_Dataset/dry_bean_train.csv")
test_df = pd.read.csv("Homework_b7/Dry_Bean_Dataset/dry_bean_test.csv")
train_df.head()
train_df.shape
train_df.columns

# Chia X và y
X_train = train_df.drop("class", axis=1)
y_train = train_df["class"]
X_test = test_df.drop("class", axis=1)
y_test = test_df["class"]
print(X_train.shape)
print(X_test.shape)
```

## Huấn luyện Decision Tree.

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)
dt_model.fit(X_train, y_train)

# Dự đoán.
dt_pred = dt_model.predict(X_test)

# Accuracy.
dt_acc = accuracy_score(y_test, dt_pred)
print("Accuracy của Decision Tree là:", dt_acc)
print(classification_report(y_test, dt_pred))
```

## Confusion Matrix của Decision Tree.

```python
cm_dt = confusion_matrix(y_test, dt_pred)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_dt,
    annot=True,
    fmt="d",
    cmap="Blues"
)
plt.title("Confusion Matrix của Decision Tree")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Mô hình đơn giản, dễ trực quan hóa và thời gian huấn luyện nhanh, phù hợp các bài toán có kích thước dữ liệu vừa và nhỏ. Nhưng mô hình bị overfitting khi cây phát triển quá sâu, làm giảm khả năng tổng quát hóa trên dữ liệu mới. 

## Random Forest.

```python
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
rf_model.fit(X_train, y_train)

# Dự đoán.
rf_pred = rf_model.predict(X_test)

# Accuracy.
rf_acc = accuracy_score(y_test, rf_pred)
print("Accuracy của Random Forest là:", rf_acc)
print(classification_report(y_test, rf_pred))
```

## Confusion Matrix của Random Forest.

```python
cm_rf = confusion_matrix(y_test, rf_pred)
plt.figure(figsize=(6,4))
sns.heatmap(
    cm_rf,
    annot=True,
    fmt="d",
    cmap="Reds"
)
plt.title("Confusion Matrix của Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

Mô hình kết hợp nhiều cây quyết định giúp mô hình học được nhiều đặc trưng của dữ liệu hơn và giảm ảnh hương của nhiễu. Mô hình hạn chế được hiện tượng overfitting nhờ lấy mẫu ngẫu nhiên và bỏ phiếu giữa các cây.

```python
print("Decision Tree Accuracy :", dt_acc)
print("Random Forest Accuracy:", rf_acc)
```

Random Forest có Accuracy cao hơn so với Decision Tree do hạn chế được overfitting, mang lại hiệu suất phân loại tốt hơn.
