```python
import matplotlib.pyplot as plt
import numpy as np

# Bật chế độ hiển thị trực tiếp trong notebook
%matplotlib inline

# Tạo dữ liệu test đơn giản
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Vẽ biểu đồ
plt.plot(x, y, label="Đường Sin")
plt.title("Biểu đồ Test Output")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()

# Hiển thị
plt.show()
```
